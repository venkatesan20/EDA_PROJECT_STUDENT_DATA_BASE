```python
sns.countplot(x='Age',data=df)
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()
```

```python
sns.histplot(df['Age'],kde=True,color='r')
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()
```

```python
sns.boxplot(x='Absences',data=df)
plt.title('Absences Distribution')
plt.xlabel('Absences')
plt.ylabel('Distribution')
plt.show()
```

```python
num_df=df.select_dtypes(include=['int64','float64'])
```

```python

plt.figure(figsize=(10,6))

sns.heatmap(num_df.corr(),annot=True)
plt.show()
```

```python
sns.lineplot(x='Age',y='Absences',data=df,color='g')
plt.title('Age vs G1')
plt.show()
```

```python
sns.scatterplot(x='Age',y='Absences',data=df,color='r',markers='o')
plt.title('Age vs Absences')
plt.show()
```
